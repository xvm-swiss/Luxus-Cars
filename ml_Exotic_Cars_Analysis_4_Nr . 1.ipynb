# Distribute all cars according to origin country

```python
data["Car Make"].unique()
```

```python
cars_country_list = []

for car in data['Car Make'] :

    if car in [ 'Porsche', 'Audi', 'BMW', 'Mercedes-Benz', 'Mercedes-AMG',  ] :
        cars_country_list.append( 'German' )

    elif car in [ 'Lamborghini', 'Ferrari', 'Maserati', 'Alfa Romeo', 'Pagani', 'Pininfarina'] :
         cars_country_list.append( 'Italian' ) 

    elif car in ["McLaren", "Aston Martin", "Jaguar", "Lotus", "Bentley","Rolls-Royce", "Ariel", "TVR", "Ultima"]:  
        cars_country_list.append( 'UK' )  
    
    elif car in ["Chevrolet", "Ford", "Dodge", "Tesla", "Shelby"]:  
        cars_country_list.append( 'US' )   

    elif car in ["Nissan", "Lexus", "Acura", "Mazda", "Toyota", "Subaru"]:  
        cars_country_list.append( 'Japan' )  

    elif car in ["Bugatti", "Alpine"]:  
        cars_country_list.append( 'France' )     

    elif car in ["Koenigsegg", "Polestar"]:  
        cars_country_list.append( 'Sweden' )  

    elif car in [ "Rimac"]:  
        cars_country_list.append( 'Croatia' )  

    elif car in ["Kia"]:  
        cars_country_list.append( 'South Korea' )  

    elif car in ["W Motors"]:  
        cars_country_list.append( 'United Arab Emirates' )  

    else :
       cars_country_list.append( 'Other' )        
        
```

```python
data.insert(2, "Country", cars_country_list)
```

# --> PDF Exotic Cars Analysis Assignment


```python
New_Segment = []

for price in data['Price (in USD)']:
    if price < 70000:
        New_Segment.append('Lower')
    
    elif price < 120000:
        New_Segment.append('Lower Mainstream')

    elif price < 180000:
        New_Segment.append('Mainstream')

    elif price < 250000:
        New_Segment.append('Upper Mainstream')

    elif price < 400000:
        New_Segment.append('Premium')

    elif price < 800000:
        New_Segment.append('Luxury')
    else:
        New_Segment.append('Exotic')
```

```python
data.insert(4, 'New_Segment', New_Segment)
```
