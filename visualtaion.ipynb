```python
import pandas as pd
import plotly.express as px

# Daten in ein Dictionary packen
data = {
    "Land": ["Deutschland", "UK", "Italien", "USA", "Japan", "Frankreich", "Schweden", "Kroatien", "UAE", "Südkorea"],
    "Anzahl": [German_car, UK_car, Italian_car, US_car, Japan_car, France_car, Sweden_car, Croatia_car, UAE_car, South_Korea_car]
}

# Temporäres DataFrame erstellen
df_plot = pd.DataFrame(data)

# Balkendiagramm erstellen
fig = px.bar(df_plot, 
             x='Land', 
             y='Anzahl', 
             color='Land',
             color_discrete_sequence=px.colors.qualitative.Safe,
             title='Anzahl der Autos nach Herkunftsland',
             text_auto=True) # Zeigt die Werte direkt über den Balken an

fig.show()
```

```python

# 1. Definition der Grenzen (Bins) und Labels
bins = [0, 250, 500, 750, 1000, np.inf]
labels = ['bis 250 HP', '251-500 HP', '501-750 HP', '751-1000 HP', '> 1000 HP']

# 2. Neue Spalte für die Gruppen erstellen
df['HP_Group'] = pd.cut(df['Horsepower'], bins=bins, labels=labels)


# Zählen und Spalten sauber benennen
df_grouped = df['HP_Group'].value_counts().reset_index()
df_grouped.columns = ['HP_Group', 'Count']

# Visualisierung
fig = px.bar(
    df_grouped, 
    x="HP_Group", 
    y="Count",
    color="HP_Group",
    title='Anzahl der Fahrzeuge nach PS-Klasse',
    category_orders={"HP_Group": labels},
    labels={'Count': 'Anzahl Fahrzeuge', 'HP_Group': 'PS-Klasse'} # Schöne Achsenbeschriftung
)


# 1. Sicherstellen, dass die Spalte numerisch ist
# 'coerce' wandelt Fehler (z.B. Text wie "unbekannt") in NaN (leere Werte) um
df['Horsepower'] = pd.to_numeric(df['Horsepower'], errors='coerce')

# Optional: Zeilen ohne PS-Wert entfernen, falls gewünscht
# df = df.dropna(subset=['Horsepower'])

# 2. Deine Bins und Labels
bins = [0, 250, 500, 750, 1000, float('inf')]
labels = ['bis 250 HP', '251-500 HP', '501-750 HP', '751-1000 HP', '> 1000 HP']

# 3. Jetzt funktioniert pd.cut
df['HP_Group'] = pd.cut(df['Horsepower'], bins=bins, labels=labels)

# 4. Zählen
df_grouped = df['HP_Group'].value_counts().reset_index()

fig.show()
```
